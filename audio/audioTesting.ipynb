```python
import pygame
import os
```

```python
def playMusic(folder, songName):
    songDir = os.path.join(folder, songName)

    if not os.path.exists(songDir): print(f"File {songName} does not exist."); return

    pygame.mixer.music.load(songDir)
    pygame.mixer.music.play()

    print(f"Now playing: {songName[:-4]}")
```

```python
try:
    pygame.mixer.init()
except pygame.error as E:
    print(f"Error occured during initialization:\n{E}")\
    
audio_dir = "audio_files"

if not os.path.isdir(audio_dir): print(f"Error accessing audio files. {audio_dir} does not exist.")

m4a_files = [file for file in os.listdir(audio_dir) if file.endswith(".mp3")]

if not m4a_files: print(f"No files located in {audio_dir}")

for index, file in enumerate(m4a_files, start=1):
    print(f"{index}. {file}")

choice = int(input("What song number do you want to play? ")) - 1
playMusic(audio_dir, m4a_files[choice])
```
